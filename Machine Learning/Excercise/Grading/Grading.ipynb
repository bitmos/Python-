```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Grading.xlsx',header=2)
df.drop(["Unnamed: 4","Unnamed: 5","Unnamed: 6","Unnamed: 7"],axis=1,inplace=True)
df
```

```python
df['Grade'] = df['Grade'].replace('?', np.nan)
df
```

```python
df1=df[41:]
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.Grade=le.fit_transform(dfle.Grade)
dfle
```

```python
x= dfle[['Percentage']]
x
```

```python
y=dfle.Grade
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
X_train
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
cross_val_score(LogisticRegression(),X_train,y_train)
```

```python
from sklearn.svm import SVC
cross_val_score(SVC(),X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
cross_val_score(LinearRegression(),X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
cross_val_score(RandomForestClassifier(),X_train,y_train)
```

```python
from sklearn.cluster import KMeans
cross_val_score(KMeans(),X_train,y_train)
```

```python
model=SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
for i in df1['Percentage']:
    df1.loc[df1.Percentage == i, 'Grade']=model.predict([[i]])
```

```python
df1
```

```python
df=pd.concat([dfle,df1])
df
```
