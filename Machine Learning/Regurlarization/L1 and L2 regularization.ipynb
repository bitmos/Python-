```python
import pandas as pd
df=pd.read_csv('housing.csv')
df.head()
```

```python
df.nunique()
```

```python
df.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
```

```python
df.isna().sum()
```

```python
df=df[cols_to_use]
df.head()
```

```python
cols_to_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[cols_to_zero]=df[cols_to_zero].fillna(0)
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
x=df.drop('Price',axis=1)
y=df["Price"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
xtrain
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
from sklearn import linear_model
lasso=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso.fit(xtrain,ytrain)
```

```python
lasso.score(xtest,ytest)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge.fit(xtrain,ytrain)
ridge.score(xtest,ytest)
```
