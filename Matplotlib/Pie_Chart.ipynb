```python
import matplotlib.pyplot as plt

slices = [59219, 55466, 47544, 36443, 35917,
          31991, 27097, 23030, 20524, 18523,
          18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL',
          'Python', 'Java', 'Bash/Shell/PowerShell',
          'C#', 'PHP', 'C++', 'TypeScript', 'C',
          'Other(s):', 'Ruby', 'Go', 'Assembly']
slices = slices[0:5]
labels = labels[0:5]
explode = [0,0.025,0.05,0.075,0.1]

plt.style.use('classic')
plt.pie(slices, labels = labels, explode=explode,
        shadow = True,startangle = 90,
        wedgeprops={'edgecolor':'white'},autopct='%1.1f%%')
plt.show()
```
